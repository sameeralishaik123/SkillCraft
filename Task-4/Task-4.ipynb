```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# For better visuals
plt.style.use('seaborn-v0_8')
sns.set(rc={'figure.figsize':(10,6)})
```

```python
df = pd.read_csv("dft-road-casualty-statistics-casualty-provisional-mid-year-unvalidated-2022 (1).csv")
df.head()
```

```python
df.info()
df.describe()
df.isnull().sum()
```

```python
df.info()
df.describe()
df.isnull().sum()
```

```python
# Drop duplicates if any
df.drop_duplicates(inplace=True)

# Handle missing values if present
df = df.dropna()
```

```python
sns.countplot(data=df, x='casualty_severity', palette='coolwarm')
plt.title('Distribution of Casualty Severity')
plt.xlabel('Casualty Severity (1 = Fatal, 2 = Serious, 3 = Slight)')
plt.ylabel('Count')
plt.show()
```

```python
sns.boxplot(data=df, x='sex_of_casualty', y='age_of_casualty')
plt.title('Age vs Sex of Casualties')
plt.xlabel('1 = Male, 2 = Female')
plt.ylabel('Age of Casualty')
plt.show()
```

```python
top_hotspots = df['lsoa_of_casualty'].value_counts().head(10)
sns.barplot(x=top_hotspots.values, y=top_hotspots.index, palette='viridis')
plt.title('Top 10 Accident Hotspots')
plt.xlabel('Number of Accidents')
plt.ylabel('LSOA Location Code')
plt.show()
```

```python
sns.countplot(data=df, x='casualty_home_area_type', hue='casualty_severity', palette='Set2')
plt.title('Casualty Severity by Home Area Type (1=Urban, 2=Rural)')
plt.xlabel('Home Area Type')
plt.ylabel('Count')
plt.show()
```

```python
sns.barplot(data=df, x='casualty_imd_decile', y='casualty_severity', ci=None, palette='mako')
plt.title('Average Severity by IMD Decile (1=Most Deprived, 10=Least Deprived)')
plt.xlabel('IMD Decile')
plt.ylabel('Average Severity')
plt.show()
```

```python
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```
